```python
# Importing the datasets

from google.colab import drive
drive.mount('/content/drive')

df1 = pd.read_csv('/content/drive/MyDrive/EDA Data/application_data.csv')
df1.head(5)
```

```python
from google.colab import drive
drive.mount('/content/drive')

df2 = pd.read_csv('/content/drive/MyDrive/EDA Data/previous_application.csv')
df2.head(5)
```

```python
# Combining Previous Application and Current Application Dataset

df = df1.merge(df2, on='SK_ID_CURR', how='left')
print(df.head(2))
```

```python
df.shape
```

**Cleaning the Data**

```python
# Checking Null Values
df.isnull().sum()
```

```python
# Finding percentage of null values
missing_percent = df.isnull().sum()*100/df.shape[0]
missing_percent.sort_values(ascending=False)
```

```python
#Removing columns where more than 75% of data is missing

df = df.loc[:, df.isnull().mean() < 0.75]
df.head(5)
```

```python
df.shape
```

```python
# Comparing data from overlapping columns

for col in df.columns:
    if col.endswith('_x'):
        base = col[:-2]  # remove '_x'
        match_col = base + '_y'
        if match_col in df.columns:
            print(f"\nComparing {col} and {match_col}:")
            print((df[col] == df[match_col]).value_counts())
            print((df[col] == df[match_col]).value_counts()/df.shape[0]*100)
```
